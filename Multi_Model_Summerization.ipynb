```python
import pandas as pd

df = pd.read_csv("/content/Input_Text.csv")
texts = df["original_text"].tolist()
domains = df["domain"].tolist()
text_ids = df["text_id\t\t"].tolist()
```

```python
models = {
    "M1": "facebook/bart-large-cnn",
    "M2": "t5-base",
    "M3": "google/pegasus-cnn_dailymail",
    "M4": "allenai/led-base-16384",
    "M5": "sshleifer/distilbart-cnn-12-6"
}
```

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
import time
import torch

summaries = []

device_str = "cuda" if torch.cuda.is_available() else "cpu"

for model_id, model_name in models.items():
    tokenizer = AutoTokenizer.from_pretrained(model_name)
    model = AutoModelForSeq2SeqLM.from_pretrained(model_name).to(device_str)

    for text_id, domain, text in zip(text_ids, domains, texts):
        start = time.time()

        if "t5" in model_name:
            input_text = "summarize: " + text
        else:
            input_text = text

        inputs = tokenizer(
            [input_text],
            max_length=1024,
            truncation=True,
            return_tensors="pt"
        ).to(device_str)

        summary_ids = model.generate(
            inputs["input_ids"],
            max_length=130,
            min_length=40,
            num_beams=4,
            early_stopping=True
        )

        summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)

        end = time.time()

        summaries.append({
            "text_id": text_id,
            "domain": domain,
            "model": model_id,
            "summary": summary,
            "inference_time": round(end - start, 3)
        })
```

```python
summary_df = pd.DataFrame(summaries)
summary_df.to_csv("summaries.csv", index=False)
```

# Evaluating Scores

```python
import pandas as pd
import numpy as np

from rouge_score import rouge_scorer
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
texts_df = pd.read_csv("/content/Input_Text.csv")
summaries_df = pd.read_csv("/content/summaries.csv")

text_map = dict(zip(texts_df["text_id\t\t"], texts_df["original_text"]))
```

```python
rouge = rouge_scorer.RougeScorer(
    ["rouge1", "rouge2", "rougeL"], use_stemmer=True
)

semantic_model = SentenceTransformer("all-MiniLM-L6-v2")
```

```python
results = []

for _, row in summaries_df.iterrows():
    original = text_map[row["text_id"]]
    summary = row["summary"]

    # ROUGE scores
    rouge_scores = rouge.score(original, summary)

    rouge1 = rouge_scores["rouge1"].fmeasure
    rouge2 = rouge_scores["rouge2"].fmeasure
    rougeL = rouge_scores["rougeL"].fmeasure

    # Semantic similarity
    orig_emb = semantic_model.encode(original)
    summ_emb = semantic_model.encode(summary)
    semantic_sim = cosine_similarity(
        [orig_emb], [summ_emb]
    )[0][0]

    # Compression ratio
    compression = len(summary.split()) / len(original.split())

    results.append({
        "text_id": row["text_id"],
        "domain": row["domain"],
        "model": row["model"],
        "rouge1": round(rouge1, 4),
        "rouge2": round(rouge2, 4),
        "rougeL": round(rougeL, 4),
        "semantic_similarity": round(semantic_sim, 4),
        "compression_ratio": round(compression, 4),
        "inference_time": row["inference_time"]
    })
```

```python
eval_df = pd.DataFrame(results)
```

```python
final_scores = (
    eval_df
    .groupby(["domain", "model"])
    [["rouge1", "rouge2", "rougeL", "semantic_similarity", "compression_ratio", "inference_time"]]
    .mean()
    .reset_index()
)

final_scores.to_csv("evaluation_scores.csv", index=False)
```
